```python
import pandas as pd

# Carregar o conjunto de dados
file_path = "Youtube_dataset_all_dataScience_channels.csv"  # Altere para o caminho correto do seu arquivo
df = pd.read_csv(file_path)

# Exibir as primeiras linhas do DataFrame
print("Primeiras linhas do arquivo CSV:")
print(df.head())

# 1. Mostrar as duplicatas por coluna
print("\nDuplicatas em cada coluna:")

# Verificar duplicatas em cada coluna individualmente
for column in df.columns:
    # Exibir as duplicatas da coluna (se houver)
    duplicates = df[df.duplicated(subset=[column])]
    if not duplicates.empty:
        print(f"\nDuplicatas na coluna '{column}':")
        print(duplicates)
    else:
        print(f"\nNão há duplicatas na coluna '{column}'.")

# 2. Remover duplicatas em todas as colunas e mostrar o número de duplicatas removidas
print("\nRemovendo duplicatas...")

# Verificar o número total de duplicatas no DataFrame
duplicatas = df.duplicated().sum()
print(f"\nNúmero total de linhas duplicadas: {duplicatas}")

# Remover as duplicatas do DataFrame
df_cleaned = df.drop_duplicates()

# 3. Remover valores nulos
print("\nRemovendo valores nulos...")

# Verificar e remover linhas com valores nulos
nulos = df_cleaned.isnull().sum().sum()  # Número total de valores nulos
print(f"Número total de valores nulos antes da remoção: {nulos}")

# Remover as linhas com valores nulos
df_cleaned = df_cleaned.dropna()

# Exibir o número de valores nulos após a remoção
nulos_restantes = df_cleaned.isnull().sum().sum()
print(f"Número total de valores nulos após a remoção: {nulos_restantes}")

# Exibir o número de linhas restantes após a remoção de duplicatas e nulos
print(f"Linhas restantes após a remoção de duplicatas e nulos: {len(df_cleaned)}")

# Exibir as primeiras linhas do DataFrame limpo
print("\nPrimeiras linhas do DataFrame após a remoção de duplicatas e nulos:")
print(df_cleaned.head())

# Salvar o DataFrame limpo (opcional)
df_cleaned.to_csv("Youtube_dataset_cleaned.csv", index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Gráfico de Distribuição do Número de Visualizações
plt.figure(figsize=(10, 6))
sns.histplot(df['Views'], bins=50, color='skyblue')
plt.title("Distribuição do Número de Visualizações")
plt.xlabel("Views")
plt.ylabel("Frequência")
plt.show()
```

```python
# Agrupar por canal e somar as visualizações
top_channels = df.groupby('Channel_Name')['Views'].sum().sort_values(ascending=False).head(10)

# Visualizar o resultado
plt.figure(figsize=(12, 6))
top_channels.plot(kind='bar', color='green')
plt.title("Top 10 Canais com Mais Visualizações")
plt.ylabel("Número de Visualizações")
plt.show()
```

```python
# Matriz de correlação
correlation = df[['Views', 'Like_count', 'Comment_Count']].corr()

# Visualizar a correlação
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlação entre Visualizações, Curtidas e Comentários")
plt.show()
```
